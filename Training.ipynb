```python
from deep_emotion_recognition import DeepEmotionRecognizer
# initialize instance
# inherited from emotion_recognition.EmotionRecognizer
# default parameters (LSTM: 128x2, Dense:128x2)
deeprec = DeepEmotionRecognizer(emotions=['angry','happy','neutral','sad'], n_rnn_layers=2, n_dense_layers=2, rnn_units=128, dense_units=128)
# train the model
deeprec.train()
# get the accuracy
#print(deeprec.test_score())
# predict angry audio sample
print(deeprec.predict_proba("data/validation/Actor_10/03-02-05-02-02-02-10_angry.wav"))
```
